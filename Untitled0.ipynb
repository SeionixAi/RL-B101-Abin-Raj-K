Even or Odd

```python
num = int(input("Enter a number: "))
if num % 2 == 0:
  print("Even number")
else:
  print("Odd number")
```

```python
numbers = [10, 20, 30, 40, 50]

print("Sum:", sum(numbers))
print("Max:", max(numbers))
print("Min:", min(numbers))
```
